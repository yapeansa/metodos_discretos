## 1) Descubra os seguintes polinômios de Taylor, plote e aproxime os seguintes valores:




### a) $f(x)=\sqrt[7]{x}$ obter $P_1(x)$ (linear). Qual o gráfico e o valor de $\sqrt[7]{1.1}$ usando $P_1(x)$? Use $a=1$.

Para a função $f(x) = \sqrt[7]{x}$, temos $f'(x) = \frac{1}{7\,\sqrt[7]{x^{6}}}$. Segue que $f(1) = 1$ e $f'(1) = \frac{1}{7}$. Desta forma, o polinômio de Taylor de $f$ em torno de $a = 1$ é dado por $P_{1}(x) = 1 + \frac{1}{7}\,(x - 1)$. No algoritmo a seguir implementamos a função para a representar o polinômio $P_{1}$ e efetuar o cálculo do valor de $\sqrt[7]{1.1}$, obtendo como resultado o valor $P_{1}(1.1) = 1.0142857142857142 \approx \sqrt[7]{1.1}$. O gráfico de $P_{1}(x)$ também é plotado juntamente com o gráfico da funço $f$.

```python
import numpy as np
import matplotlib.pyplot as plt
from math import factorial


# Função f dada pelo problema
def f(x):
    return x**(1/7)


# Polinômio de Taylor da função f
def P_1(x):
    return (1 + (1/7)*(x - 1))


# Avaliação no ponto 1.1 usando o polinômio
print(f'Polinômio avaliado no ponto 1.1: P_1(1.1) = {P_1(1.1)}.')
print(f'Função avaliada no ponto 1.1: f(1.1) = {f(1.1)}.')

# Plotando os gráficos

fig = plt.figure(figsize=(8, 8))

# Gerando 100 pontos de 0 até 2
x = np.linspace(0.5, 1.5, 100)
# Avaliando o polinômio nos pontos de x
p_1 = P_1(x)
# Avaliando f nos pontos de x
func = f(x)
# Gráfico do polinômio
plt.plot(x, p_1, c='r', label='$P_{1}(x) = 1 + \\frac{1}{7}\\,(x - 1)$')
# Gráfico da função f
plt.plot(x, func, c='b', ls='--', label='$f(x) = \\sqrt[7]{x}$')

plt.title('Gráficos de f(x) e P$_{1}$(x)')
plt.xlabel('x')
plt.grid(alpha=0.6)
plt.legend()
plt.show()
```

### b)  $f(x)=e^{x}$ obter $P_2(x)$ (quadrático). Qual o gráfico e o valor de $e^{0.2}$ usando $P_2(x)$? Use $a=0$.

Primeiramente, observamos que para $f(x) = e^{x}$, temos $f'(x) = f''(x) = e^{x}$ para todo $x \in \mathbb{R}$. Desta forma, segue que $f(0) = f'(0) = f''(0) = 1$. Logo, o polinômio de Taylor de ordem 2 em torno de $a = 0$ da função $f$ é dado por:
$$P_{2}(x) = 1 + x + \frac{x^{2}}{2}.$$
A seguir, temos a implementação do algoritmo para a representação do polinômio $P_{2}(x)$ para o cálculo do valor de $e^{0.2}$, de modo que foi obtido o valor $P_{2}(0.2) = 1.22 \approx e^{0.2}$. O gráfico de $P_{2}(x)$ também é plotado juntamente com o gráfico da função $f$.

```python
#Função f dada pelo problema
def f(x):
    return np.exp(x)


# Polinômio de Taylor para a função f
def P_2(x):
    return (1 + x + (x**2)/2)


# Avaliação do polinômio no ponto 0.2
print(f'Polinômio avaliado no ponto 0.2: P_2(0.2) = {P_2(0.2)}.')
print(f'Função f avaliada no ponto 0.2: f(0.2) = {f(0.2)}.')

x = np.linspace(-1, 1, 100)
p_2 = P_2(x)
func = f(x)

# Potando os gráficos
plt.figure(figsize=(8, 8))
plt.plot(x, p_2, c='r', label='$P_{2}(x) = 1 + x + \\frac{x^{2}}{2}$')
plt.plot(x, func, c='b', ls='--', label='$f(x) = e^{x}$')

plt.title('Gráficos de $f(x)$ e $P_{2}(x)$')
plt.xlabel('x')
plt.grid(alpha=0.6)
plt.legend()
plt.show()
```

## 2) Produza uma fórmula geral para o polinômio de Taylor expandido em torno de $a = 0$ de grau $n$, além disso tomando $n=20$ gere os gráficos dos polinômio obtidos.



### a) f(x) = $\frac{1}{1-x}$

Para $f(x) = \frac{1}{1 - x}$, temos $f'(x) = \frac{1}{(1 - x)^{2}}$, $f''(x) = \frac{2}{(1 - x)^{3}}$, $f'''(x) = \frac{2\,\cdot\,3}{(1 - x)^{4}}$, $f^{(4)} = \frac{2\,\cdot\,3\,\cdot\,4}{(1 - x)^{5}}$, $\ldots$, e assim por diante. Além disso, observe que $f$ não está definida para $x = 1$. Generalizando a derivada k-ésima, segue que $f^{(k)}(x) = \frac{k!}{(1 - x)^{k+1}}$ para todo $k$ inteiro não negativo, onde $f^{(0)}(x) = f(x)$. Desta forma, $f^{(k)}(0) = k!$, de modo que a fórmula geral para o polinômio de Taylor de grau $n$, em torno de $a = 0$, da função $f$ é dado por $P_{n}(x) = \sum\limits_{k = 0}^{n} \frac{f^{k}(0)}{k!}\,x^{k} = \sum\limits_{k = 0}^{n} \frac{k!}{k!}\,x^{k} = \sum\limits_{k = 0}^{n} x^{k}$. Agora, tomando $n = 20$, a implementação em Python para o polinômio e os respectivos gráficos são apresentados a seguir.

```python
# Função f dada pelo problema
def f(x):
    return 1/(1 - x)


def P_n(x, n):
    p = 0

    for k in range(n+1):
        p += x**k

    return p


x = np.linspace(-1, 0.9, 100)
n = 20  # Tomando n = 20
p_20 = P_n(x, n)
func = f(x)

# Plotando os gráficos
plt.figure(figsize=(8, 8))
plt.plot(x, p_20, c='r', label='$P_{20}(x)$')
plt.plot(x, func, c='b', ls='--', label='$f(x)$')

plt.title('Gráficos de $f(x)$ e $P_{20}(x)$')
plt.xlabel('x')
plt.grid(alpha=0.6)
plt.legend()
plt.show()
```


### b) f(x) = $\sqrt{1+x}$

Para a função $f(x) = \sqrt{1+x}$, temos $f'(x) = \frac{1}{2\,\sqrt{1 + x}}$, $f''(x) = \frac{-1}{2^{2}\,\sqrt{(1 + x)^{3}}}$, $f'''(x) = \frac{3}{2^{3}\,\sqrt{(1 + x)^{5}}}$, $f^{(4)}(x) = \frac{-3\,\cdot\,5}{2^{4}\,\sqrt{(1+x)^{7}}}$, $f^{(5)}(x) = \frac{3\,\cdot\,5\,\cdot\,7}{2^{5}\,\sqrt{(1+x)^{9}}}$, e assim por diante. Além disso, observe que $f$ está definida para $x \geq -1$. Generalizando a derivada k-ésima, temos $f^{(k)}(x) = \frac{(-1)^{k+1}\,[1\,\cdot\,3\,\cdot\,5\,\ldots\,(2k -5)\,\cdot\,(2k - 3)]}{2^{k}\,\sqrt{(1+x)^{2k - 1}}}$ para todo $k$ inteiro maior que $1$. Logo, $f^{(k)}(0) = \frac{(-1)^{k+1}\,[1\,\cdot\,3\,\cdot\,5\,\ldots\,(2k - 3)]}{2^{k}}$ para todo $k \geq 1$ inteiro e $f(0) = 1$. Agora, observe que $1\,\cdot\,3\,\cdot\,5\,\ldots\,(2k - 3) = \frac{(2k -2)!}{2^{k-1}\,(k-1)!}$ para todo $k \geq 1$ inteiro. Isto pode ser demonstrado pela técnica matemática conhecida como o Princípio da Indução Finita. De fato, para $k = 1$, temos $\frac{(2(1) - 2)!}{2^{1 - 1}\,(1 - 1)!} = 1$. Para $k = 2$, temos $\frac{(2(2) - 2)!}{2^{2 - 1}\,(2 - 1)!} = 1$. Agora, suponhamos por hipótese de indução que $1\,\cdot\,3\,\cdot\,5\,\ldots\,(2k - 3) = \frac{(2k -2)!}{2^{k-1}\,(k-1)!}$ é verdadeiro, onde $k > 1$ é inteiro positivo arbitrário. Então, vamos provar que a relação vale para $k+1$. De fato,
\begin{align*}
    1\,\cdot\,3\,\cdot\,5\,\ldots\,(2k - 3)\,\cdot\,(2(k+1) - 3) &= \frac{(2k -2)!}{2^{k-1}\,(k-1)!}\,(2(k+1) - 3) \\
    &= \frac{(2k -2)!}{2^{k-1}\,(k-1)!}\,(2k - 1) = 2\,k\,\frac{(2\,k - 2)!}{2^{k}\,k!}\,(2k - 1) \\
    &= \frac{2k\,(2k - 1)\,(2k - 2)!}{2^{k}\,k!} = \frac{(2k)!}{2^{k}\,k!} = \frac{(2(k + 1) - 2)!}{2^{(k+1) - 1}\,((k+1) - 1)!}.
\end{align*}
Portanto, pelo Princípio da Indução Matemática, $1\,\cdot\,3\,\cdot\,5\,\ldots\,(2k - 3) = \frac{(2k -2)!}{2^{k-1}\,(k-1)!}$ para todo $k \geq 1$ inteiro. Obtemos então uma expressão geral para a derivada k-ésima:
$$f^{(k)}(0) = \frac{(-1)^{k+1}}{2^{k}}\,\frac{(2\,k - 2)!}{2^{k-1}\,(k-1)!} = \frac{(-1)^{k+1}\,(2\,k - 2)!}{2^{(2k-1)}\,(k-1)!}.$$
Diante disso, a forma geral do polinômio de Taylor de ordem $n$ em torno de $a = 0$ da função é dada por:
$$P_{n}(x) = 1 + \sum\limits_{k = 1}^{n} \frac{(-1)^{k+1}\,(2\,k - 2)!}{2^{(2k-1)}\,(k-1)!}\,\frac{x^{k}}{k!}.$$
A seguir, toma-se $n = 20$ e plota-se os gráficos de $P_{20}(x)$ e da função $f(x)$.

```python
# Função f dada pelo problema
def f(x):
    return np.sqrt(1 + x)


# Polinômio de Taylor da função f
def P_n(x, n):
    p = 1  # Primeiro termo do polinômio

    for k in range(1, n+1):
        # Derivada k-ésima no ponto a = 0
        f_n = (((-1)**(k+1))*factorial((2*k)-2))/((2**(2*k-1))*factorial(k-1))
        p += f_n*(x**k/factorial(k))

    return p


x = np.linspace(-1, 1.3, 100)
n = 20  # Tomando n = 20
p_20 = P_n(x, n)
func = f(x)

# Plotando os gráficos
plt.figure(figsize=(8, 8))
plt.plot(x, p_20, c='r', ls='-', label='$P_{20}(x)$')
plt.plot(x, func, c='b', ls='--', label='$f(x)$')

plt.title('Gráficos de $f(x)$ e $P_{20}(x)$')
plt.xlabel('x')
plt.grid(alpha=0.6)
plt.legend()
plt.show()
```

## 3) Encontre e plote os seguintes polinômios de Taylor, além disso calcule os limitantes superiores de erro como se pede:




### a) Encontre um limitante superior para o erro ao aproximar $f(x) = e^x$, para $x \in [-1, 1]$, pelo polinômio de Taylor de grau 3 expandido em torno de $a = 0$. Calcule o erro absoluto de $P_3(x=1)$ com $f(x=1)$ e compare com o limitante obtido

Para a função $f(x) = e^{x}$, temos $f'(x) = f''(x) = f'''(x) = e^{x}$ para todo $x \in \mathbb{R}$. Assim, $f(0) = f'(0) = f''(0) = f'''(0) = 1$. Diante disso, o polinômio de Taylor de grau 3 expandido em torno de $a = 0$ é dado por $P_{3}(x) = 1 + x + \frac{x^{2}}{2} + \frac{x^{3}}{3!}$. Além disso, $f^{(4)}(x) = e^{x}$. Utilizando o resíduo, temos o seguinte erro:
$$R(x) = \frac{x^{4}}{4!}\,e^{\eta} = \frac{x^{4}}{24}\,e^{\eta}, \text{ para algum } \eta \in [-1, 1].$$
Como a função exponencial é crescente, segue que ao aproximarmos $e^{1}$ pelo polinômio $P_{3}(x)$, o limitante superior para o erro é dado por:
$$|R(x)| \leq \max\,\left|\frac{x^{4}}{24}\,e^{\eta}\right| \leq \left|\frac{1^{4}}{24}\,e^{1}\right| = \frac{e}{24}.$$
Pelo algoritmo a seguir, o limitante superior obtido é `lim_sup` $= \frac{e}{24} = 0.11326174285246021$. Ainda pelo algoritmo a seguir, o erro absoluto obtido ao aproximar $e^{1}$ pelo polinômio $P_{3}(x)$ tem o valor `erro_abs` $ = |P_{3}(1) - f(1)| = 0.05161516179237857$, o que está de acordo com o limitante superior obtido `erro_abs` $ \leq $ `lim_sup`.

```python
# Função f dada pelo problema
def f(x):
    return np.exp(x)


# Polinômio de Taylor da função f
def P_3(x):
    return (1 + x + x**2/2 + x**3/6)


x = np.linspace(-1, 1, 100)
p_3 = P_3(x)
func = f(x)

# Plotando os gráficos
plt.figure(figsize=(8, 8))
plt.plot(x, p_3, c='r', label='$P_{3}(x)$')
plt.plot(x, func, c='b', ls='--', label='$f(x)$')

plt.title('Gráficos de $f(x)$ e $P_{3}(x)$')
plt.xlabel('x')
plt.grid(alpha=0.6)
plt.legend()
plt.show()

# Cálculo do limitante superior para o erro
lim_sup = np.e/24
a = P_3(1)
b = f(1)
# Cálculo do erro absoluto
erro_abs = np.abs(a - b)

print(f'Polinômio avaliado em x = 1: P_3(1) = {a}.')
print(f'Função f avaliada em x = 1: f(1) = {b}.')
print(f'Erro absoluto: {erro_abs}.')
print(f'Limitante superior obtido: {lim_sup}.')
```

### b) Encontre o polinômio de Taylor de grau 2 para a função $f(x)=e^x sin(x)$ em torno do ponto $a = 0$. Determine um limitante superior do erro para essa aproximação para $x \in [-\pi/4, \pi/4]$. Calcule o erro absoluto de $P_2(x=0)$ com $f(x=0)$ e compare com o limitante obtido

Como $f(x) = e^{x}\,sin(x)$, temos $f'(x) = e^{x}\,(cos(x) + sin(x))$. Da mesma forma, temos $f''(x) = e^{x}\,(cos(x) + sin(x)) + e^{x}\,(cos(x) - sin(x)) = 2\,e^{x}\,cos(x)$. Com isso, temos $f(0) = 0$, $f'(0) = 1$ e $f''(0) = 2$. Assim, o polinômio de Taylor de grau 2 para a função $f$ em torno do ponto $a = 0$ é dado por $P_{2}(x) = x + x^{2}$. Agora, observamos que $f'''(x) = 2\,(e^{x}\,cos(x) - e^{x}\,sin(x)) = 2\,e^{x}\,(cos(x) - sin(x))$. Usando o resíduo, temos:
$$R(x) = \frac{x^{3}}{6}\,2\,e^{\eta}\,(\cos(\eta) - \sin(\eta)), \text{ para algum } \eta \in \left[-\frac{\pi}{4}, \frac{\pi}{4}\right].$$
Definindo, $\mu(\eta) = 2\,e^{\eta}\,(\cos(\eta) - \sin(\eta))$, temos $\mu'(\eta) = 2\,e^{\eta}\,(\cos(\eta) - \sin(\eta)) + 2\,e^{\eta}\,(-\sin(\eta) - \cos(\eta)) = -4\,e^{\eta}\,\sin(\eta)$. Assim, $\mu'(\eta) = 0$ quando $\sin(\eta) = 0$, isto é, $\eta = 0$, pois $\eta \in \left[-\frac{\pi}{4}, \frac{\pi}{4}\right]$. Logo, a função $\mu$ atinge seu valor máximo em $\eta = 0$.

Diante disto, ao aproximarmos $e^{\frac{\pi}{4}}\,\sin(\frac{\pi}{4})$ pelo polinômio $P_{2}(x)$, obtemos o seguinte limitante superior para o erro:
\begin{align*}
    |R(x)| \leq \max \left|\frac{x^{3}}{6}\,2\,e^{\eta}\,(\cos(\eta) - \sin(\eta)) \right| \leq \left|\frac{(\frac{\pi}{4})^{3}}{6}\,2\,e^{0}\,(\cos(0) - \sin(0)) \right| = \frac{\pi^{3}}{192}.
\end{align*}
Efetuando os cálculo na implementação do algoritmo a seguir, obtemos o seguinte valor para o limitante superior: `lim_sup` $= 0.16149102437656154$. O Erro absoluto obtido para $P_{2}\left(\frac{\pi}{4}\right)$ com $f\left(\frac{\pi}{4}\right)$ é `erro_abs` $= 0.1486347584524923$, o que está de acordo com o valor para o limitante superior para o erro pois `erro_abs` $\leq$ `lim_sup`.

```python
# Função f dada pelo problema
def f(x):
    return np.exp(x)*np.sin(x)


# Polinômio de Taylor da função f
def P_2(x):
    return (x + x**2)


x = np.linspace(-np.pi/4, np.pi/4, 100)
p = P_2(x)
func = f(x)

# Plotando os gráficos
plt.figure(figsize=(8, 8))
plt.plot(x, p, c='r', label='$P_{2}(x)$')
plt.plot(x, func, c='b', ls='--', label='$f(x)$')

plt.title('Gráficos de $f(x)$ e $P_{2}(x)$')
plt.xlabel('x')
plt.grid(alpha=0.6)
plt.legend()
plt.show()

# Cálculo do limitante superior para o erro
lim_sup = (np.pi**3)/192
a = P_2(np.pi/4)
b = f(np.pi/4)
# Cálculo do erro absoluto
erro_abs = np.abs(a - b)

print(f'Polinômio avaliado em 0: P_2(0) = {a}.')
print(f'Função avaliada em 0: f(0) = {b}.')
print(f'Erro absoluto: {erro_abs}.')
print(f'Limitante superior obtido: {lim_sup}.')
```

## 4) Encontre o polinômio de Taylor de $2^a$ ordem, $P_2$ de $f(x,y)= e^{2x+3y}$ próximo a $(0,0)$. Além disso, plote o polinômio obtido junto da função $f(x,y)$. Aproxime o valor de (x,y)=(0.5, 0.1) e calcule o erro absoluto.

Para a função $f(x, y) = e^{2\,x + 3\,y}$, temos as seguintes derivadas parciais: $f_{x}(x, y) = 2\,e^{2\,x + 3\,y}$, $f_{y}(x, y) = 3\,e^{2\,x + 3\,y}$, $f_{xy}(x, y) = 6\,e^{2\,x + 3\,y}$, $f_{xx}(x, y) = 4\,e^{2\,x + 3\,y}$, $f_{yy}(x, y) = 9\,e^{2\,x + 3\,y}$. Logo, $f(0, 0) = 1$, $f_{x}(0, 0) = 2$, $f_{y}(0, 0) = 3$, $f_{xy}(0, 0) = 6$, $f_{xx}(0, 0) = 4$ e $f_{yy}(0, 0) = 9$. Então, o polinômio de Taylor de 2ª ordem da função $f$ em torno de $(0, 0)$ é dado por:
$$P_{2}(x, y) = 1 + 2\,x + 3\,y + 6\,x\,y + 2\,x^{2} + \frac{9}{2}\,y^{2}.$$
Na implementação de código a seguir, são plotados os gráficos de $P_{2}(x, y)$ e de $f(x, y)$. O valor obtido para o polinômio avaliado no ponto $(0.5, 0.1)$ é $P_{2}(0.5, 0.1) = 3.1449999999999996$ e o erro absoluto obtido tem o valor `erro_abs` $= 0.5242966676192449$.

```python
# Função f(x, y) dada pelo problema
def f(x, y):
    return np.exp(2*x + 3*y)


# Polinômio de Taylor para a função f
def P_2(x, y):
    return (1 + 2*x + 3*y + 6*x*y + 2*x**2 + (9/2)*y**2)


x = np.linspace(-1, 1, 30)
y = np.linspace(-1, 1, 30)
X, Y = np.meshgrid(x, y)
p = P_2(X, Y)
func = f(X, Y)

fig = plt.figure(figsize=(10, 10))
ax = fig.add_subplot(projection='3d')
ax.plot_wireframe(X, Y, func, color='b', label='$f(x, y)$')
ax.plot_wireframe(X, Y, p, color='r', label='$P_{2}(x, y)$')
plt.title('Gráficos de $f(x, y)$ e $P_{2}(x, y)$', fontweight='bold')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.view_init(20, -45)

plt.legend()
plt.show()

a, b = 0.5, 0.1
w = P_2(a, b)
z = f(a, b)
print(f'Polinômio avaliado no ponto (0.5, 0.1): P_2(0.5, 0.1) = {w}.')
print(f'Função avaliada no ponto (0.5, 0.1): f(0.5, 0.1) = {z}.')

erro_abs = np.abs(w - z)
print(f'Erro absoluto: {erro_abs}.')
```
