```python
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle

# Criação da figura e dos eixos
fig, ax = plt.subplots()

# Definição dos pontos do quadrado de lado 0.1
x_coords = [0, 0.1, 0.1, 0, 0]
y_coords = [0, 0, 0.1, 0.1, 0]

# Desenho do quadrado externo com um zorder alto para que ele fique por cima
# O valor 3 é arbitrário, mas maior que o padrão dos eixos
ax.plot(x_coords, y_coords, color='red', linewidth=2, zorder=3)

# Configuração dos limites dos eixos para um melhor enquadramento
ax.set_xlim(-0.01, 0.12)
ax.set_ylim(-0.01, 0.12)

# Adição dos pontos e suas coordenadas
ax.plot(0, 0, 'o', color='black', markersize=3)
ax.text(0.005, 0.005, '(0,0)', ha='left', va='bottom')

ax.plot(0.1, 0, 'o', color='black', markersize=3)
ax.text(0.105, 0.005, '(0.1, 0)', ha='left', va='bottom')

ax.plot(0.1, 0.1, 'o', color='black', markersize=3)
ax.text(0.105, 0.105, '(0.1, 0.1)', ha='left', va='bottom')

ax.plot(0, 0.1, 'o', color='black', markersize=3)
ax.text(0.005, 0.105, '(0, 0.1)', ha='left', va='bottom')

# Adição das letras A, B, C e D
ax.text(-0.005, 0.05, 'A', ha='right', va='center', color='blue', fontsize=16)
ax.text(0.05, 0.105, 'B', ha='center', va='bottom', color='blue', fontsize=16)
ax.text(0.105, 0.05, 'C', ha='left', va='center', color='blue', fontsize=16)
ax.text(0.05, -0.005, 'D', ha='center', va='top', color='blue', fontsize=16)

# Ocultar os números dos eixos
ax.set_xticklabels([])
ax.set_yticklabels([])

# Configurações para manter os eixos x e y
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_position('zero')
ax.spines['bottom'].set_position('zero')

# Posicionar os labels 'x' e 'y' manualmente
ax.text(0.115, -0.005, 'x', ha='left', va='center', fontsize=12)
ax.text(-0.005, 0.115, 'y', ha='center', va='bottom', fontsize=12)

# Desenhar o quadrado preenchido
# É uma boa prática dar um zorder mais alto para o quadrado interno também
inner_square = Rectangle((0.04, 0.04), 0.02, 0.02, facecolor='orange', zorder=4)
ax.add_patch(inner_square)

ax.set_aspect('equal')
plt.tight_layout()
plt.savefig('dominio.png')
plt.show()
```
